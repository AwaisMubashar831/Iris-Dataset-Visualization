### **1. Project Overview**
Analyze the classic Iris dataset to:
- Understand feature distributions
- Compare flower species characteristics
- Identify most distinguishing features
- Visualize multivariate relationships

### **2. Dataset Description**
Contains 150 samples of 3 Iris species:
- Setosa, Versicolor, and Virginica
- Four features:
  - Sepal length (cm)
  - Sepal width (cm)
  - Petal length (cm)
  - Petal width (cm)

```python
### **3. Setup & Import**

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
# Styling
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

4. **Data Loading**

```python
# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

# Check first 5 rows
df.head()
```

5. **Exploratory Data Analysis**

```python
# Basic statistics
print("Dataset Summary:")
print(df.describe())

# Class distribution
print("\nSpecies Distribution:")
print(df['species'].value_counts())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

6. **Data Visualization**

**6.1 Pair Plot (Multivariate Analysis)**

```python
sns.pairplot(df, hue='species', palette='viridis')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

**6.2 Box Plots (Feature Distribution)**

```python
plt.figure(figsize=(12,8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='species', y=feature, data=df)
    plt.title(f'{feature} Distribution')
plt.tight_layout()
plt.show()
```

**6.3 Violin Plots (Density Distribution)**

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='species', y='petal length (cm)', data=df, inner='quartile')
plt.title("Petal Length Distribution by Species")
plt.show()
```

**6.4 Scatter Plot (Most Discriminating Features)**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)',
                hue='species', data=df, palette='viridis',
                s=100, edgecolor='black')
plt.title("Petal Length vs Petal Width")
plt.show()
```

**6.5 Correlation Heatmap**

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')
plt.title("Feature Correlation Matrix")
plt.show()
```

7. **Conclusions**

Clear Species Separation: Petal measurements show distinct clusters

Key Differentiator: Petal length is most distinguishing feature

Strong Correlations:

Petal length & width (0.96)

Sepal length & petal length (0.87)

Setosa Distinctiveness: Easily separable from other species

Virginica Characteristics: Largest petals on average
